**Classes**

````python
# What is the output of the following code?
class Owlbear:
    def __init__(self, name):
        self.name = name
        self.mood = "confused"
        self.__hp = 100

    def secret_method(self):
        def inner_method():
            print("A wise old owl never tells.")
        inner_method()

    @classmethod
    def rename(cls, new_name):
        cls.name = new_name

    @staticmethod
    def hoot():
        print("Hoot! Hoot!")
    
    @property
    def roar(self):
        return f"{self.name} lets out a thunderous hoot-roar!"

    @property
    def hitPoints(self):
        return self.__hp

    @hitPoints.setter
    def hitPoints(self, new_hp):
        if new_hp >= 0:
            self.__hp = new_hp
        else:
            print("HP cannot be negative!")


class OwlbearChild(Owlbear):
    def __init__(self, name):
        super().__init__(name)
        self.mood = "playful"

ob = Owlbear("Hoot")
bb = OwlbearChild("Hooty")

````

````python
def sharpened_daggers(damage_list):
    result = []
    for dmg in sorted(damage_list, reverse=True)[:-1] + [damage_list[0] + 1]:
        if dmg ** 2 <= 20:
            result.append(dmg ** 2)
        else:
            result.append(20)

    return result

damage_list = [1, 5, 10]
print(sharpened_daggers(damage_list))
````
